# Part 2: KS test

```
from __future__ import print_function, division
import pandas as pd
import pylab as pl
import scipy as sp
from scipy import stats
%pylab inline
pl.style.use("https://raw.githubusercontent.com/fedhere/DSPS/master/fbb.mplstyle")
```

1. The null hypothesis that is being tested during the KS test is as follows, The two samples do not come from the same distribution.

2. The 'statistic' or 'pivotal quantity' for the KS test is d=max*|C1(x)-C2(x)| where C1 and C2 are representative of the cumulative distributions of x for sample one and sample 2. d is representative of the maximum difference between the two cumulative sample distributions, hence the max part of the statistic.  This pivotal quantity is how the KS tests accpets or rejects the hypothesis telling us if it is likely that the samples are part of the same parent distribution or not.

3. The probability distribution of this statistic depends on the absolute value of the difference between the two cumulative distributions and whether or not they are part of the same parent distribution. The probability distrubution also depends on the selected p-value.
