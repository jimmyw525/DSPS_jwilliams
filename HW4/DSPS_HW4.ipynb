```
import numpy as np
import matplotlib.pyplot as plt
%pylab inline

def square(side):
  ax = plt.figure(figsize=(5,5)).add_subplot(1,1,1)
  plt.scatter(0,0)
  plt.scatter(0,side)
  plt.scatter(side,0)
  plt.scatter(side,side)
  plt.plot((-side,-side),(-side,side))
  plt.plot((side,-side),(-side,-side))
  plt.plot((side,side),(-side,side))
  plt.plot((side,-side),(side,side))
  
  return ax
```
